```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load the solution
data = np.loadtxt("/Users/danieltompkins/Documents/advanced-programming/solution.txt")
nx, ny = 100, 100
nt = 500
u = data.reshape((nt, nx, ny))
```

```python
# Create a grid for x and y coordinates
x = np.linspace(0, 10, nx)  # Domain size in x (0 to 10)
y = np.linspace(0, 10, ny)  # Domain size in y (0 to 10)
X, Y = np.meshgrid(x, y)

# Plot the solution at the final time step as a surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, u[-1], cmap='viridis')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Amplitude')
ax.set_title('2D Wave Equation Solution at t = final')

# Add a color bar
fig.colorbar(surf, ax=ax, label='Amplitude')

# Show the plot
plt.show()
```
